```python
y = dataFrame["Type"].values
X= dataFrame.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
scaler.fit(X_train)
```

```python
scaler.transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(units=32,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=16,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=16,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=16,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=X_train,y=y_train,epochs=200,validation_data=(X_test,y_test),verbose=1)
```

```python
modelLoss= pd.DataFrame(model.history.history)
```

```python
modelLoss.plot()
```

```python
model = Sequential()

model.add(Dense(units=32,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=16,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=16,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=16,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=200,validation_data=(X_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot()
```

```python
predict = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
predicted_classes = (predict > 0.5).astype(int)
print(classification_report(y_test, predicted_classes))
```

```python
predict_classes = np.round(predict).astype(int)
print(confusion_matrix(y_test, predict_classes))
```
