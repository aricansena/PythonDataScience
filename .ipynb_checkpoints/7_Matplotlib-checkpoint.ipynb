```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [10,20,30,30,30,40,50,60,70,75]
```

```python
weightList = [20,60,80,85,86,87,70,90,95,90]
```

```python
npAgeList = np.array(ageList)
npWeightList = np.array(weightList)
```

```python
plt.plot(npAgeList,npWeightList,"g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Compare Age vs Weight")
```

```python
npArray1 = np.linspace(0,10,20)
```

```python
npArray1
```

```python
npArray2 = npArray1 ** 2
```

```python
npArray2
```

```python
plt.plot(npArray1,npArray2,"r*-")
```

```python
plt.subplot(1,2,1)
plt.plot(npArray1,npArray2,"r--")
plt.subplot(1,2,2) #1 row 2 columns second graph
plt.plot(npArray2,npArray1,"g--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.5,0.5])
figureAxes.plot(npArray1,npArray2,"g")
figureAxes.set_xlabel("X Ekseni")
figureAxes.set_ylabel("Y Ekseni")
```

```python
figure2 = plt.figure()
axis1 = figure2.add_axes([0.1,0.1,0.9,0.9])
axis2 = figure2.add_axes([0.2,0.5,0.3,0.3])
axis1.plot(npArray1,npArray2,"g")
axis2.plot(npArray2,npArray1,"r")
```

```python
plt.subplot()
```

```python
(myFigure,axis1)=plt.subplots(nrows=1,ncols=2)
for axis in axis1:
    axis.plot(npArray1,npArray2,"r")
plt.tight_layout()
```

```python
newFigure = plt.figure(figsize=(4,4),dpi=150)
newAxis = newFigure.add_axes([0.1,0.1,0.9,0.9])
newAxis.plot(npArray1,npArray1**2,label="**2")
newAxis.plot(npArray1,npArray1**3,label="**3")
newAxis.legend(loc=9)
```

```python
newFigure.savefig("myFigure.png",dpi=150)
```

## Visuality

```python
(myFigurei,myAxis)=plt.subplots()
myAxis.plot(npArray1,npArray2,color="#4A99B1")
myAxis.plot(npArray2,npArray1,color="#CA99B1")
```

```python
(myFigurei,myAxis)=plt.subplots()
myAxis.plot(npArray1,npArray1+6,color="pink",linewidth=5,linestyle="--")
myAxis.plot(npArray1,npArray1+4,color="purple",linewidth=3,linestyle=":")
myAxis.plot(npArray1,npArray1+2,color="black",linewidth=0.7)

myAxis.plot(npArray1,npArray1+4,color="purple",linewidth=3,linestyle="--")
```
